```python
import numpy as np
import pandas as pd

RAND_MAX = 32_767
Time = 0.007
```

```python
np.random.randint(0, RAND_MAX)
```

```python
def gent_data(point_num: int, size: int, k: int):
  np.random.seed(42)
  size = 3
  
  param_r = 1.9
  param_R = 1
  scale = 130
  data = []
  true_data = []

  for i in range(point_num):
    U = np.random.randint(0, RAND_MAX) * (0.03 / RAND_MAX);
    V = np.random.randint(0, RAND_MAX) * (0.03 / RAND_MAX);
    e = np.sqrt(-2.0 * np.log(U)) * np.sin(2.0 * 3.14 * V);
    tmp = []
    true_tmp = []

    for j in range(size):
      if (j % size == 0):
        tmp.append(e + scale * ((param_R - param_r) * np.cos(param_r / param_R * (1 + Time*k)) + param_r / param_R * np.cos((1 + Time*k) - param_r / param_R * (1 + Time*k))))
        true_tmp.append(scale * ((param_R - param_r) * np.cos(param_r / param_R * (1 + Time*k)) + param_r / param_R * np.cos((1 + Time*k) - param_r / param_R * (1 + Time*k))))
      elif (j % size == 1):
        tmp.append(e + scale * ((param_R - param_r) * np.sin(param_r / param_R * (1 + Time*k)) - param_r / param_R * np.sin((1 + Time*k) - param_r / param_R * (1 + Time*k))))
        true_tmp.append(scale * ((param_R - param_r) * np.sin(param_r / param_R * (1 + Time*k)) - param_r / param_R * np.sin((1 + Time*k) - param_r / param_R * (1 + Time*k))))
      else:
        tmp.append(e + scale * ((param_R - param_r) * np.sin(param_r / param_R * (1 + Time*k)) - param_r / param_R * np.cos((1 + Time*k) - param_r / param_R * (1 + Time*k))))
        true_tmp.append(scale * ((param_R - param_r) * np.sin(param_r / param_R * (1 + Time*k)) - param_r / param_R * np.cos((1 + Time*k) - param_r / param_R * (1 + Time*k))))
    
    data.append(tmp)
    true_data.append(true_tmp)

  return data, true_data
```

```python
iter = int(19/Time)
point_num = 10_000
size = 3

datas = []
datas_true = []

for k in range(iter):
  data, data_true = gent_data(point_num, size, k)
  # datas += data
  # datas_true += data_true 
  df = pd.DataFrame(data)
  dtf = pd.DataFrame(data_true)
  
  df.to_csv(f'data_{point_num}_i_{k:06d}.csv')
  dtf.to_csv(f'data_true_{point_num}_i_{k:06d}.csv')
```

```python
df = pd.DataFrame(datas)
dtf = pd.DataFrame(datas_true)
```
